## 1.4 Extra — String Methods

_Continue here after finishing the core 1.4 notebook._

There are many string methods which allow us to format strings.

- **capitalize()**: Converts the first character to capital letter
- **count()**: Returns occurrences of substring in string

```python
challenge = 'thirty days of python'
print(challenge.count('y'))        # 3
print(challenge.count('y', 7, 14)) # 1
print(challenge.count('th'))       # 2
```

- **endswith()**: Checks if a string ends with a specified ending

```python
challenge = 'thirty days of python'
print(challenge.endswith('on'))   # True
print(challenge.endswith('tion')) # False
```

- **find()**: Returns the index of the first occurrence of a substring, if not found returns -1
- **rfind()**: Returns the index of the last occurrence
- **index()** / **rindex()**: Same as find/rfind but raises `ValueError` if not found

```python
challenge = 'thirty days of python'
sub_string = 'da'
print(challenge.find(sub_string))    # 7
print(challenge.find('xyz'))         # -1
print(challenge.index(sub_string))   # 7
print(challenge.rindex('on', 8))     # 19
```

- **format()**: Formats string into a nicer output

```python
first_name = 'Alex'
last_name = 'Park'
age = 250
job = 'teacher'
country = 'Finland'
sentence = 'I am {} {}. I am a {}. I am {} years old. I live in {}.'.format(
    first_name, last_name, job, age, country)
print(sentence)
```

### Checking Methods

| Method | Description |
|---|---|
| `isalnum()` | True if all chars are alphanumeric |
| `isalpha()` | True if all chars are alphabetic |
| `isdecimal()` | True if all chars are decimal (0-9) |
| `isdigit()` | True if all chars are digits |
| `isnumeric()` | True if all chars are numeric |
| `isidentifier()` | True if valid variable name |
| `islower()` | True if all alphabet chars are lowercase |
| `isupper()` | True if all alphabet chars are uppercase |

```python
print('ThirtyDays'.isalnum())    # True
print('ThirtyDays'.isalpha())    # True
print('123'.isdecimal())         # True
print('30'.isdigit())            # True
print('\u00BD'.isnumeric())      # True (½)
print('my_var'.isidentifier())   # True
print('hello'.islower())         # True
print('HELLO'.isupper())         # True
```

### Transform & Search Methods

```python
# join
web_tech = ['HTML', 'CSS', 'JavaScript', 'React']
print(' '.join(web_tech))            # HTML CSS JavaScript React
print('# '.join(web_tech))           # HTML# CSS# JavaScript# React

# strip
challenge = 'thirty days of pythoonnn'
print(challenge.strip('noth'))       # irty days of py

# replace
challenge = 'thirty days of python'
print(challenge.replace('python', 'coding'))  # thirty days of coding

# split
print(challenge.split())             # ['thirty', 'days', 'of', 'python']
print('a, b, c'.split(', '))         # ['a', 'b', 'c']

# title / swapcase / startswith
print(challenge.title())             # Thirty Days Of Python
print(challenge.swapcase())          # THIRTY DAYS OF PYTHON
print(challenge.startswith('thirty'))# True
```
