## Tasks

1. **Variables**: Create a list called `expenses` containing 7 integers (your spending for Monday to Sunday).
2. **Operations**:
   - Calculate `total_spent` using the `sum()` function on the list.

   - Find the `max_spent` and `min_spent` using `max()` and `min()`.
3. **Strings**: 
   - Create a summary string using an f-string: `"Total: <total>, Avg: <avg>"` (round average to 2 decimals).

4. **Lists**: Create a new list `above_average` containing only the expenses that were higher than the `daily_average`.

Implement your logic in the code cell below. Use `assert` statements to verify your results at the end.

```python
# 1. Your data
expenses = [120, 45, 300, 80, 150, 400, 50]

# 2. Your Calculations
total_spent = sum(expenses)
daily_average = total_spent / 7
max_spent = max(expenses)
min_spent = min(expenses)

# 3. String formatting
summary = f"Total: {total_spent}, Avg: {daily_average:.2f}"
peak_day_index = expenses.index(max_spent)
peak_msg = f"Peak Spending: Day {peak_day_index + 1}"

# 4. List logic
above_average = [e for e in expenses if e > daily_average]

# --- Tests ---
assert total_spent == 1145
assert round(daily_average, 2) == 163.57
assert max_spent == 400
assert min_spent == 45 # wait, min is 45 in the list above
assert len(above_average) == 2 # 300, 400
assert summary == "Total: 1145, Avg: 163.57"

print("All tasks completed successfully!")
```
