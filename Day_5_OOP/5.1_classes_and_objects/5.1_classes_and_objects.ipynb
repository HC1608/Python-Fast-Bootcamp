## Submodule 5.1 — Classes and Objects

Everything in Python is an object. A **class** is a blueprint for creating objects. An **object** is an instance of a class.

### Creating a Class

We use the `class` keyword to define a class. By convention, class names use **PascalCase**.

```python
class Person:
    pass

print(Person)  # <class '__main__.Person'>
```

### Creating an Object (Instantiation)

An object is created by calling the class like a function.

```python
p = Person()
print(p)  # <__main__.Person object at 0x...>
```

### The `__init__` Constructor

The `__init__` method is called automatically when a new object is created. It initializes the object's attributes.

- `self` refers to the current instance of the class.

```python
class Person:
    def __init__(self, first_name, last_name, age, country):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.country = country

p = Person('Harsh', 'Choudhary', 25, 'India')
print(p.first_name)  # Harsh
print(p.age)          # 25
```

### Instance Methods

Methods are functions defined inside a class. They describe the behaviors of an object.

```python
class Person:
    def __init__(self, first_name, last_name, age, country):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.country = country

    def full_name(self):
        return f'{self.first_name} {self.last_name}'

    def greet(self):
        return f'Hello, my name is {self.full_name()}. I am {self.age} years old.'

p = Person('ABC', 'DEF', 25, 'India')
print(p.full_name())  # ABC DEF
print(p.greet())      # Hello, my name is ABC DEF. I am 25 years old.
```

### Default Parameter Values

You can provide default values for constructor parameters.

```python
class Person:
    def __init__(self, first_name='John', last_name='Doe', age=0, country='Unknown'):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.country = country

p1 = Person()
print(p1.first_name)  # John

p2 = Person('Alice', 'Smith', 30, 'USA')
print(p2.first_name)  # Alice
```

### The `__str__` Method

The `__str__` method defines the string representation of an object (used by `print()`).

```python
class Person:
    def __init__(self, first_name, last_name, age, country):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.country = country

    def __str__(self):
        return f'Person({self.first_name} {self.last_name}, age={self.age})'

p = Person('ABC', 'DEF', 25, 'India')
print(p)  # Person(ABC DEF, age=25)
```

### Class vs Instance Attributes

- **Class attributes** are shared by all instances.
- **Instance attributes** are unique to each instance.

```python
class Dog:
    species = 'Canis familiaris'  # class attribute

    def __init__(self, name, age):
        self.name = name  # instance attribute
        self.age = age

dog1 = Dog('Buddy', 3)
dog2 = Dog('Max', 5)

print(dog1.species)  # Canis familiaris
print(dog2.species)  # Canis familiaris
print(dog1.name)     # Buddy
print(dog2.name)     # Max
```
