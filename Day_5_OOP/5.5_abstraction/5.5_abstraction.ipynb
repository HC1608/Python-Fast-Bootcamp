## Submodule 5.5 — Abstraction

**Abstraction** means hiding complex implementation details and showing only the essential features. In Python, we achieve this with **Abstract Base Classes (ABCs)** using the `abc` module.

### Why Abstraction?

- Enforces a contract: child classes **must** implement certain methods.
- You cannot instantiate an abstract class directly.
- Makes large codebases more maintainable.

### Abstract Base Classes

Use `ABC` as the parent class and `@abstractmethod` to declare methods that must be overridden.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Cannot instantiate an abstract class
try:
    s = Shape()
except TypeError as e:
    print(f'Error: {e}')
```

### Implementing Abstract Classes

A child class **must** implement all abstract methods, or it remains abstract too.

```python
import math
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

r = Rectangle(10, 5)
c = Circle(7)
print(f'Rectangle — area: {r.area()}, perimeter: {r.perimeter()}')
print(f'Circle    — area: {c.area():.2f}, perimeter: {c.perimeter():.2f}')
```

### Incomplete Implementation

If a child class doesn't implement all abstract methods, Python will raise `TypeError` when you try to instantiate it.

```python
class IncompleteShape(Shape):
    def area(self):
        return 0
    # perimeter() is NOT implemented

try:
    s = IncompleteShape()
except TypeError as e:
    print(f'Error: {e}')
```

### Abstract Classes with Concrete Methods

Abstract classes can also have regular (concrete) methods that provide shared logic.

```python
class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass

    def describe(self):  # concrete method — shared by all
        return f'I am {self.name}. {self.speak()}'

class Dog(Animal):
    def speak(self):
        return 'Woof!'

class Cat(Animal):
    def speak(self):
        return 'Meow!'

d = Dog('Buddy')
c = Cat('Whiskers')
print(d.describe())  # I am Buddy. Woof!
print(c.describe())  # I am Whiskers. Meow!
```

### Real-World Example: Payment System

Abstraction is ideal for defining contracts in systems like payment processors.

```python
class PaymentProcessor(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

    @abstractmethod
    def refund(self, amount):
        pass

class CreditCardPayment(PaymentProcessor):
    def pay(self, amount):
        return f'Paid ₹{amount} via Credit Card.'

    def refund(self, amount):
        return f'Refunded ₹{amount} to Credit Card.'

class UPIPayment(PaymentProcessor):
    def pay(self, amount):
        return f'Paid ₹{amount} via UPI.'

    def refund(self, amount):
        return f'Refunded ₹{amount} via UPI.'

def process_payment(processor, amount):
    print(processor.pay(amount))

cc = CreditCardPayment()
upi = UPIPayment()
process_payment(cc, 500)
process_payment(upi, 200)
```
