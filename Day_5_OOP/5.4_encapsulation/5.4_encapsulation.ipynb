## Submodule 5.4 — Encapsulation

**Encapsulation** is the practice of bundling data and methods that operate on that data within a class, and restricting direct access to some of the object's internals. This prevents accidental modification and enforces controlled access.

### Public, Protected, and Private Attributes

Python uses naming conventions (not strict enforcement) to signal access levels:

| Prefix | Convention | Access |
|---|---|---|
| `name` | Public | Accessible from anywhere |
| `_name` | Protected | Should only be accessed within the class and subclasses |
| `__name` | Private | Name-mangled; hard to access from outside |

```python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner          # public
        self._account_type = 'Savings'  # protected
        self.__balance = balance     # private

acc = Account('Harsh', 1000)
print(acc.owner)          # Harsh — accessible
print(acc._account_type)  # Savings — accessible but discouraged

# This will raise an AttributeError:
try:
    print(acc.__balance)
except AttributeError as e:
    print(f'Error: {e}')
```

### Name Mangling

Private attributes with `__` prefix are name-mangled to `_ClassName__attribute`. You _can_ access them, but you shouldn't.

```python
# Name mangling — not recommended for regular use
print(acc._Account__balance)  # 1000
```

### Getters and Setters

Instead of directly accessing private data, we provide controlled methods.

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f'Deposited {amount}. New balance: {self.__balance}'
        return 'Deposit amount must be positive.'

    def withdraw(self, amount):
        if amount > self.__balance:
            return 'Insufficient funds.'
        if amount <= 0:
            return 'Withdrawal amount must be positive.'
        self.__balance -= amount
        return f'Withdrew {amount}. New balance: {self.__balance}'

acc = BankAccount('Harsh', 5000)
print(acc.get_balance())   # 5000
print(acc.deposit(1000))   # Deposited 1000. New balance: 6000
print(acc.withdraw(2000))  # Withdrew 2000. New balance: 4000
print(acc.withdraw(10000)) # Insufficient funds.
```

### The `@property` Decorator

The Pythonic way to create getters and setters. Access attributes like normal while keeping validation logic.

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius  # calls the setter

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError('Temperature below absolute zero is not possible.')
        self._celsius = value

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

t = Temperature(25)
print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0

t.celsius = 100
print(t.fahrenheit)   # 212.0

# This raises ValueError:
try:
    t.celsius = -300
except ValueError as e:
    print(e)
```

### Read-Only Properties

Define only a getter (no setter) to make an attribute read-only.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @property
    def area(self):
        import math
        return math.pi * self._radius ** 2

c = Circle(5)
print(c.radius)        # 5
print(f'{c.area:.2f}') # 78.54

try:
    c.radius = 10  # AttributeError — read-only
except AttributeError as e:
    print(f'Error: {e}')
```
