## Submodule 5.2 — Inheritance

**Inheritance** allows a class (child) to acquire the attributes and methods of another class (parent). It promotes code reuse and models _"is-a"_ relationships.

### Basic Inheritance

A child class inherits from a parent class by passing the parent in parentheses.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} makes a sound.'

class Dog(Animal):
    pass

d = Dog('Buddy')
print(d.speak())  # Buddy makes a sound.
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} makes a sound.'

class Dog(Animal):
    def bark(self):
        return f'{self.name} says Woof!'

d = Dog('Buddy')
print(d.speak())  # Buddy makes a sound.
print(d.bark())   # Buddy says Woof!
```

The `Dog` class inherits `__init__` and `speak()` from `Animal` without writing any code.

### Overriding Methods

A child class can **override** a parent method by redefining it.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} makes a sound.'

class Dog(Animal):
    def speak(self):
        return f'{self.name} barks!'

class Cat(Animal):
    def speak(self):
        return f'{self.name} meows!'

dog = Dog('Buddy')
cat = Cat('Whiskers')
print(dog.speak())  # Buddy barks!
print(cat.speak())  # Whiskers meows!
```

### The `super()` Function

`super()` lets you call the parent class's methods from the child, which is useful when extending (not replacing) behavior.

```python
class Person:
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age

    def full_name(self):
        return f'{self.first_name} {self.last_name}'

class Student(Person):
    def __init__(self, first_name123, last_name123, age123, university):
        super().__init__(first_name123, last_name123, age123)  # call parent __init__
        self.university = university

    def info(self):
        return f'{self.full_name()} studies at {self.university}'

s = Student('Harsh', 'Choudhary', 25, 'MIT')
print(s.info())       # Harsh Choudhary studies at MIT
print(s.full_name())  # Harsh Choudhary  (inherited)
```

### `isinstance()` and `issubclass()`

- `isinstance(obj, Class)` — checks if an object is an instance of a class (or its parent).
- `issubclass(ChildClass, ParentClass)` — checks the class hierarchy.

```python
print(isinstance(s, Student))   # True
print(isinstance(s, Person))    # True  (Student IS-A Person)
print(issubclass(Student, Person))  # True
```

### Multi-Level Inheritance

A class can inherit from a class that itself inherits from another.

```python
class Animal:
    def breathe(self):
        return 'I breathe oxygen.'
    def speak(self):
        return 'I make a sound.'
    def move(self):
        return 'I can move.'

class Mammal(Animal):
    def feed_young(self):
        return 'I feed my young with milk.'
    def speak(self):
        return 'I make a mammal sound.'

class Dog(Mammal):
    def fetch(self):
        return 'I love fetching sticks!'    
    def move(self):
        return 'I run on four legs.'
    

d = Dog()
print(d.breathe())     # from Animal
print(d.feed_young())  # from Mammal
print(d.speak())       # from Mammal
print(d.fetch())       # from Dog
print(d.move())        # from Dog
```

### Multiple Inheritance

Python supports inheriting from more than one parent class.

```python
class Flyer:
    def fly(self):
        return 'I can fly!'

class Swimmer:
    def swim(self):
        return 'I can swim!'
    def fly(self):
        return 'I can\'t fly, I swim!'

class Duck( Swimmer, Flyer):
    def quack(self):
        return 'Quack!'

d = Duck()
print(d.fly())    # I can fly!
print(d.swim())   # I can swim!
print(d.quack())  # Quack!
```

### Method Resolution Order (MRO)

When multiple parents share a method, Python follows the **MRO** (left to right, depth first).

```python
print(Duck.__mro__)
# (<class 'Duck'>, <class 'Flyer'>, <class 'Swimmer'>, <class 'object'>)
```
